```python
#This could also be done by overwriting the values just with indexing
#when f = d, then f[condition]=new value
```

```python
#Populate f with d
f = d
#Update for the values in f
#Conditions 
f = np.where(((f > d_min) & (f < d_mean)),25 , f)
f = np.where(((f > d_mean) & (f < d_max)),75 , f)
f = np.where((f == d_mean),50 , f)
f = np.where((f == d_min),0 , f)
f = np.where((f == d_max),100 , f)
f
```
